```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# Load the data from the CSV file
df = pd.read_csv('input_data_NormalDistribution.csv')
# Extract values from the DataFrame
data = df['value']

# Fit a normal distribution to the data
mu, std = norm.fit(data)

# Plot the histogram
plt.hist(data, bins=10, density=True, alpha=0.6, color='g')

# Plot the PDF (Probability Density Function) of the fitted normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)

# Add labels and a title
plt.title("Fit results: mu = %.2f,  std = %.2f" % (mu, std))
plt.xlabel('Value')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data for a normal distribution
mu = 0
sigma = 1
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Normal Distribution', color='blue')

# Shade the areas under the curve for 1, 2, and 3 standard deviations
plt.fill_between(x, y, where=[(xi >= mu - sigma) and (xi <= mu + sigma) for xi in x], color='lightcoral', alpha=0.5, label='68%')
plt.fill_between(x, y, where=[(xi >= mu - 2*sigma) and (xi <= mu + 2*sigma) for xi in x], color='lightgreen', alpha=0.5, label='95%')
plt.fill_between(x, y, where=[(xi >= mu - 3*sigma) and (xi <= mu + 3*sigma) for xi in x], color='lightblue', alpha=0.5, label='99.7%')

# Add labels and legend
plt.title('68-95-99.7 Rule: Empirical Rule for a Normal Distribution')
plt.xlabel('Standard Deviations from the Mean')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
